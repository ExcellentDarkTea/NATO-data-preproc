```python
import os
import pandas as pd

DATA_DIR = "data"
FOLDERS = ["olha", "daria_z", "robert"]

events = []

for person in FOLDERS:
    person_path = os.path.join(DATA_DIR, person)

    for root, _, files in os.walk(person_path):
        for file in files:
            if file.endswith("events.csv"):   # or file == "events.csv"
                csv_path = os.path.join(root, file)

                df = pd.read_csv(csv_path)

                df["participant"] = person
                df["events_file"] = csv_path.replace("\\", "/")

                events.append(df)

events_df = pd.concat(events, ignore_index=True)

events_df
```

```python
import datetime
def parse_log_datetime(dt_str):
    # dt_str: '2026-04-15 11:03:30.647'
    return datetime.datetime.strptime(dt_str, "%Y-%m-%d %H:%M:%S.%f")
def extract_time(value):
    if pd.isna(value):
        return pd.NA
    text = str(value).strip()
    if not text:
        return pd.NA
    return text.split()[-1]
def timestamp_to_seconds(ts):
    # Handles missing values and strings like '11:07:23.359275'
    if pd.isna(ts):
        return pd.NA
    ts = str(ts).strip()
    if not ts:
        return pd.NA
    time_part, _, fraction = ts.partition(".")
    t = datetime.datetime.strptime(time_part, "%H:%M:%S")
    ms = float(f"0.{fraction}") if fraction else 0
    return t.hour * 3600 + t.minute * 60 + t.second + ms
events_df['prefix'] = events_df.iloc[:, 0].str.split('/').str[0]
#save only time without date
# events_df['time'] = events_df['timestamp'].apply(extract_time)
# events_df['Timestamp'] = events_df['time'].apply(timestamp_to_seconds)
events_df['Timestamp'] = events_df['timestamp']
events_df
```

```python
import pandas as pd
import numpy as np
```

```python
def get_participant_session(path):
    """Extract 'USER NAME' and 'SESSION NUMBER' from a path"""
    path = path.replace('\\', '/')
    parts = [p for p in path.split('/') if p]
    idx = parts.index('data')
    participant = parts[idx + 1]
    session = parts[idx + 2]
    return participant, session

def to_time_datetime(values):
    return pd.to_datetime(values, format='%H:%M:%S.%f', errors='coerce')

def standardize_series(series):
    values = pd.to_numeric(series, errors='coerce')
    mean_value = values.mean()
    std_value = values.std(ddof=0)
    if pd.isna(std_value) or std_value == 0:
        return values - mean_value
    return (values - mean_value) / std_value

def find_local_peaks(values, threshold=None):
    values = np.asarray(values, dtype=float)
    if values.size < 3:
        return np.array([], dtype=int)
    if threshold is None:
        threshold = values.mean() + 0.5 * values.std(ddof=0)
    middle = values[1:-1]
    peak_mask = (middle > values[:-2]) & (middle >= values[2:]) & (middle >= threshold)
    return np.where(peak_mask)[0] + 1

def summarize_ecg_window(window, participant, session, sensor, source_file, value_col='ECG_uV_std'):
    ecg = pd.to_numeric(window[value_col], errors='coerce').dropna().astype(float)
    if ecg.empty:
        return None

    time_values = window.loc[ecg.index, 'Timestamp_dt']
    start_time = time_values.iloc[0]
    end_time = time_values.iloc[-1]
    duration_s = (end_time - start_time).total_seconds() if pd.notna(start_time) and pd.notna(end_time) else np.nan

    ecg_values = ecg.to_numpy()
    ecg_diff = np.diff(ecg_values) if ecg_values.size > 1 else np.array([], dtype=float)
    peaks = find_local_peaks(ecg_values)

    q25 = float(np.percentile(ecg_values, 25))
    q75 = float(np.percentile(ecg_values, 75))
    label_values = window['label'].dropna()
    label = label_values.mode().iloc[0] if not label_values.empty else 'REST'
    return {
        'participant': participant,
        'session': session,
        'sensor': sensor,
        'label': label,
        'source_file': source_file,
        'window_start': start_time,
        'window_end': end_time,
        'duration_s': duration_s,
        'n_samples': int(ecg_values.size),
        'ecg_mean': float(np.mean(ecg_values)),
        'ecg_std': float(np.std(ecg_values, ddof=0)),
        'ecg_min': float(np.min(ecg_values)),
        'ecg_max': float(np.max(ecg_values)),
        'ecg_range': float(np.max(ecg_values) - np.min(ecg_values)),
        'ecg_median': float(np.median(ecg_values)),
        'ecg_q25': q25,
        'ecg_q75': q75,
        'ecg_iqr': float(q75 - q25),
        'ecg_rms': float(np.sqrt(np.mean(np.square(ecg_values)))),
        'ecg_energy': float(np.sum(np.square(ecg_values))),
        'ecg_mad': float(np.mean(np.abs(ecg_values - np.mean(ecg_values)))),
        'ecg_abs_diff_mean': float(np.mean(np.abs(ecg_diff))) if ecg_diff.size else np.nan,
        'ecg_abs_diff_std': float(np.std(np.abs(ecg_diff), ddof=0)) if ecg_diff.size else np.nan,
        'ecg_line_length': float(np.sum(np.abs(ecg_diff))) if ecg_diff.size else 0.0,
        'ecg_peak_count': int(peaks.size),
        'ecg_peak_rate': float(peaks.size / duration_s) if duration_s and duration_s > 0 else np.nan,
        'ecg_peak_mean': float(np.mean(ecg_values[peaks])) if peaks.size else np.nan,
        'ecg_peak_std': float(np.std(ecg_values[peaks], ddof=0)) if peaks.size else np.nan,
    }


# Build Start/Stop intervals per (participant, session, label)
def build_intervals(group):
    group = group.sort_values('Timestamp_dt')
    starts = group[group['event_type'] == 'Start']
    intervals = []
    for _, srow in starts.iterrows():
        stops = group[(
            (group['event_type'] == 'Stop')
            & (group['label'] == srow['label'])
            & (group['Timestamp_dt'] > srow['Timestamp_dt'])
        )]
        if not stops.empty:
            stop_ts = stops.iloc[0]['Timestamp_dt']
            intervals.append((srow['Timestamp_dt'], stop_ts, srow['label']))
    return intervals
```

# PREPROC

```python


WINDOW_SECONDS = 60
STEP_SECONDS = 10


events_df[['participant_key', 'session_key']] = events_df['events_file'].apply(
    lambda x: pd.Series(get_participant_session(x))
)
events_df['Timestamp_dt'] = to_time_datetime(events_df['Timestamp'])



intervals_by_group = {
    key: build_intervals(grp)
    for key, grp in events_df.groupby(['participant_key', 'session_key'])
}

# sliding-window features
os.makedirs('features', exist_ok=True)
ecg_feature_frames = []
labeled_signals = {}  

for participant, files in suffix_files.items():
    for f in files:
        print(f"Processing {f}...")
        p, s = get_participant_session(f)
        intervals = intervals_by_group.get((p, s), [])

        sig_df = pd.read_csv(f)
        timestamp_col = 'Packet_Timestamp'
        sig_df = sig_df.copy()
        sig_df['Timestamp_dt'] = to_time_datetime(sig_df[timestamp_col])
        sig_df['ECG_uV_std'] = standardize_series(sig_df['ECG_uV'])

        if intervals:
            starts, stops, labels = zip(*intervals)
            idx = pd.IntervalIndex.from_arrays(starts, stops, closed='both')
            positions = idx.get_indexer(sig_df['Timestamp_dt'])
            sig_df['label'] = [labels[pos] if pos != -1 else 'REST' for pos in positions]
        else:
            sig_df['label'] = 'REST'

        labeled_signals[f] = sig_df

        filename = os.path.basename(f)
        filename_no_ext = os.path.splitext(filename)[0]
        sensor = filename_no_ext[-3:]  # 'h10' or 'oh1'

        labeled_rows = sig_df[sig_df['label'].notna()].copy().sort_values('Timestamp_dt')
        if labeled_rows.empty:
            continue

        window_rows = []
        window_start = labeled_rows['Timestamp_dt'].iloc[0]
        window_end_limit = labeled_rows['Timestamp_dt'].iloc[-1] - pd.Timedelta(seconds=WINDOW_SECONDS)

        while window_start <= window_end_limit:
            window_end = window_start + pd.Timedelta(seconds=WINDOW_SECONDS)
            window = labeled_rows[(
                labeled_rows['Timestamp_dt'] >= window_start
            ) & (
                labeled_rows['Timestamp_dt'] < window_end
            )]

            if not window.empty:
                summary = summarize_ecg_window(window, p, s, sensor, f, value_col='ECG_uV_std')
                if summary is not None:
                    window_rows.append(summary)

            window_start += pd.Timedelta(seconds=STEP_SECONDS)

        features_df = pd.DataFrame(window_rows)
        if not features_df.empty:
            ecg_feature_frames.append(features_df)
            features_df.to_csv(f"features/{p}_{s}__{sensor}_ecg_features.csv", index=False)

ecg_features_df = pd.concat(ecg_feature_frames, ignore_index=True) if ecg_feature_frames else pd.DataFrame()
ecg_features_df
```
